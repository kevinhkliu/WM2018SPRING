## Read data

```python
from sklearn.externals import joblib
(tf_kp_posts,tf_kp_posts_feature_names, kp_sum_posts_clean_seg, tf_kp_comments,tf_kp_comments_feature_names, kp_sum_comments_clean_seg) = joblib.load( "result/tf_idf_kp_all.pkl" )
(tf_yao_posts,tf_yao_posts_feature_names, yao_sum_posts_clean_seg, tf_yao_comments,tf_yao_comments_feature_names, yao_sum_comments_clean_seg) = joblib.load( "result/tf_idf_yao_all.pkl" )
```

## LDA

```python
from sklearn.decomposition import LatentDirichletAllocation
import matplotlib.pyplot as plt
import time
```

```python
t0 = time.time()
lda_kp_comments = LatentDirichletAllocation(n_components=20, 
                                max_iter=20,
                                evaluate_every=200,
                                learning_method='online',
                                learning_offset = 200,
                                random_state=1)
lda_kp_comments.fit(tf_kp_comments)
print(lda_kp_comments.perplexity(tf_kp_comments))
print ("done in %0.3fs, N_iter %d, " % ((time.time() - t0), lda_kp_comments.n_iter_))

t0 = time.time()
lda_yao_comments = LatentDirichletAllocation(n_components=20, 
                                max_iter=20,
                                evaluate_every=200,
                                learning_method='online',
                                learning_offset = 200,
                                random_state=1)
lda_yao_comments.fit(tf_yao_comments)
print(lda_yao_comments.perplexity(tf_yao_comments))
print ("done in %0.3fs, N_iter %d, " % ((time.time() - t0), lda_yao_comments.n_iter_))
```

## Comments to topic

```python
import numpy as np  
comments_kp_topic_list = []
doc_topic = lda_kp_comments.transform(tf_kp_comments)
for n in range(doc_topic.shape[0]):
    topic_most_pr = doc_topic[n].argmax()
    comments_kp_topic_list.append(topic_most_pr)
    
comments_yao_topic_list = []
doc_topic = lda_yao_comments.transform(tf_yao_comments)
for n in range(doc_topic.shape[0]):
    topic_most_pr = doc_topic[n].argmax()
    comments_yao_topic_list.append(topic_most_pr)
```

## Exporting the Results

```python
joblib.dump((lda_kp_comments, comments_kp_topic_list), "result/lda_kp_all.pk")
joblib.dump((lda_yao_comments,comments_yao_topic_list), "result/lda_yao_all.pk")
```
