## KP NMF Result

```python
#NMF model
(W,H,k) = joblib.load("result/kp_nmf.pk")
tf = tf_kp_posts
tf_feature_names = tf_kp_posts_feature_names
```

```python
import numpy as np
def get_descriptor( tf_feature_names, H, topic_index, top ):
    # reverse sort the values to sort the indices
    top_indices = np.argsort( H[topic_index,:] )[::-1]
    # now get the terms corresponding to the top-ranked indices
    top_terms = []
    for term_index in top_indices[0:top]:
        top_terms.append( tf_feature_names[term_index] )
    return top_terms

for topic_index in range(k):
    descriptor = get_descriptor( tf_feature_names, H, topic_index, 20 )
    str_descriptor = ", ".join( descriptor )
    print("Topic %02d: %s" % ( topic_index+1, str_descriptor ) )
```

```python
def get_top_snippets(W, topic_index, top):
    # reverse sort the values to sort the indices
    top_indices = np.argsort( W[:,topic_index] )[::-1]
    # now get the snippets corresponding to the top-ranked indices
    top_snippets = []
    for doc_index in top_indices[0:top]:
        top_snippets.append( doc_index )
    return top_snippets
```

```python
kp_topic_posts = []
for topic_index in range(0, k):
    topic_comments = get_top_snippets( W, topic_index, 10 )
    kp_topic_posts.append(topic_comments)
    #print('The top 10 comments in topic 1')
    #for i, comment in enumerate(topic_comments):
    #   print("%02d. %s" % ( (i+1), comment ) )
print(kp_topic_posts)
```
